```python
import numpy as np
import matplotlib.pyplot as plt

# Step function definition
def step_function(x):
    return np.where(x >= 0, 1, 0)

# Input values for the task
input_values = np.array([-3.0, -1.0, 0.0, 2.0, 5.0])
output_values = step_function(input_values)

# Visualization: Plotting the step function for inputs between -5 and 5
x = np.linspace(-5, 5, 1000)
y = step_function(x)

# Plot the step function without the individual points
plt.plot(x, y, label="Step Function", color='b')
plt.title("Step Function")
plt.xlabel("Input Values")
plt.ylabel("Output Values")
plt.grid(True)
plt.axhline(0, color='black',linewidth=1)
plt.axvline(0, color='black',linewidth=1)
plt.legend()
plt.show()

# Print the output values for the given input values
print("Output Values:", output_values)
```

```python
import math

# Step function using math for scalar inputs
def step_function_scalar(x):
    return 1 if x >= 0 else 0

# Testing with scalar input
input_values = [-3.0, -1.0, 0.0, 2.0, 5.0]
output_values = [step_function_scalar(x) for x in input_values]
print("Scalar Output Values:", output_values)
```
